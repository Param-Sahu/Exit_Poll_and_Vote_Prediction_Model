# Election Exit Poll and Voter Prediction Model 
You are hired by one of the leading news channels CNBE who wants to analyze recent elections. This survey was conducted on 1525 voters with 9 variables. You have to build a model, to predict which party a voter will vote for on the basis of the given information, to create an exit poll that will help in predicting overall win and seats covered by a particular party.

## Data Dictionary for Dataset

| Feature                       | Description                                                                                                      | Interpretation                                                                                                         |
| ----------------------------- | ---------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------- |
| **`vote`**                    | The political party the voter supports. Values: `"Labour"` or `"Conservative"`                                   | This is your **target variable** — what you're trying to predict using all other features.                             |
| **`age`**                     | Age of the voter in years.                                                                                       | A numeric feature. Older or younger voters may prefer different parties.                                               |
| **`economic.cond.national`**  | Rating of the **country’s economic condition** (1–5 scale).                                                      | Subjective opinion: **1 = very bad**, **5 = very good**. May influence party preference based on national performance. |
| **`economic.cond.household`** | Rating of **personal/household financial situation** (1–5 scale).                                                | Also subjective: how well-off the respondent feels economically. Important for voter behavior.                         |
| **`Blair`**                   | How the respondent rates **Tony Blair**, leader of the Labour Party at the time (1–5 scale).                     | **1 = dislike very much**, **5 = like very much**. A high score suggests voter favors Labour.                          |
| **`Hague`**                   | Rating of **William Hague**, leader of the Conservative Party (1–5 scale).                                       | Similarly, **higher scores** suggest **preference for the Conservatives**.                                             |
| **`Europe`**                  | 11-point scale on attitude toward **European integration**. Higher = **more Eurosceptic** (less support for EU). | Political alignment with EU-related issues. Can influence support for parties based on EU stance.                      |
| **`political.knowledge`**     | How well the respondent understands each party’s EU position (0–3 scale).                                        | **0 = no knowledge**, **3 = high knowledge**. May correlate with how confidently a person votes.                       |
| **`gender`**                  | Gender of the voter: `"male"` or `"female"`.                                                                     | Used to analyze demographic influence on vote preference.                                                              |


💡 Example Interpretation:
A person who:

Rates Blair = 5, Hague = 1

Thinks national economy = 4, household = 4

Has low Euroscepticism (Europe = 2)
... is likely to vote Labour, based on ideology and leadership preference.

### Feature Scaling
📏 Is Feature Scaling Necessary?

Scaling is needed when your model is sensitive to the range of values.

| Model                          | Needs Scaling? | Why?                                           |
| ------------------------------ | -------------- | ---------------------------------------------- |
| **Logistic Regression**        | ✅ Yes          | Uses gradients and distance-based optimization |
| **KNN**                        | ✅ Yes          | Distance-based (e.g., Euclidean)               |
| **Naive Bayes**                | ❌ No           | Works with probability, not distance           |
| **LDA**                        | ✅ Yes          | Assumes Gaussian distribution; scaling helps   |
| **Random Forest / Tree-based** | ❌ No           | Not sensitive to scale                         |
| **Boosting (XGBoost, etc.)**   | ❌ Usually not  | Tree-based, doesn't care about scale           |

✅ In your dataset:

Features like age (range: 24–93) and Europe (range: 1–11) are on different scales.

**So scaling is recommended for Logistic Regression, KNN, and LDA**.

✅ Should `vote` or `gender` Be Scaled?
`vote`:
❌ No — This is your target variable, not a feature, so it should never be scaled. Leave it as 0 and 1.

`gender`:
✅ No — Even though it’s a feature, it's categorical (just coded numerically).
Scaling a binary category like gender doesn’t make sense, because:

Scaling might make values like 0.2 or 0.7, which don’t represent meaningful categories.

Models can handle 0/1 binary variables without scaling.

✅ What to Scale:
Only scale continuous or ordinal numeric features, like:

- age

- economic.cond.national

- economic.cond.household

- Blair, Hague

- Europe

- political.knowledge

| Column                | Type        | Include in Scaling? |
| --------------------- | ----------- | ------------------- |
| `vote`                | Target      | ❌ No                |
| `gender`              | Categorical | ❌ No                |
| Other numeric columns | Features    | ✅ Yes               |


```python
# Using StandardScaler for Feature Scalling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled=df_cleaned.copy()
numerical_cols = ['age', 'economic.cond.national', 'economic.cond.household', 'Blair', 'Hague', 'Europe', 'political.knowledge']
df_scaled[numerical_cols] = scaler.fit_transform(df_cleaned[numerical_cols])
```

## Model Building and Evaluation

### Logistics Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Define feature variables (X) and target variable (y)
X = df_scaled.drop(columns='vote')
y = df_scaled['vote']
```

```python
# Split the data into training and testing sets Train 70 : Test 30 (0.3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,stratify=y)
```

🔍 What is random_state in model training?
The random_state parameter is used to control the randomness of functions like train_test_split, model initialization, shuffling, etc., to make your results reproducible.

✅ Why Use random_state?
Many operations in machine learning involve randomness:
 - Splitting the data into train and test
 - Initializing model parameters (like weights)
 - Shuffling data during cross-validation or training

If you don’t fix the random seed (random_state), you'll get slightly different results each time you run the code.

🔒 Setting random_state:
By using a fixed number like random_state=42, you ensure that:
 - The same data split happens every time.
 - The model’s behavior is consistent and repeatable.
 - Others can reproduce your results exactly.

The stratify=y ensures the class distribution remains consistent between the train and test sets.

✅ Example:
Let’s say your target (y) has:
 - 70% Labour (class 1)
 - 30% Conservative (class 0)
Using stratify=y ensures both training and test sets will maintain roughly 70/30 split, so the model is trained and tested on a representative sample.

🔎 Without stratify=y, you might randomly get a test set with mostly one class, which would skew model evaluation.



```python
# Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_train_pre = model.predict(X_train)
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train,y_train_pre)
print("\nAccuracy of Trainig of model:", train_accuracy,'\n')
print("\nAccuracy of the Logistics Regression model:", accuracy,'\n')
print("\nClassification Report:\n", classification_report(y_test, y_pred),'\n')
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

✅ Interpretation:

Confusion Matrix :

 - True Positives (292): Correctly predicted Labour voters

 - True Negatives (91): Correctly predicted Conservative voters

 - False Positives (47): Predicted Labour but actually Conservative

 - False Negatives (26): Predicted Conservative but actually Labour

The model performs well on both training and test sets, showing **no overfitting**.

**It classifies Labour voters (class 1) much better than Conservative (class 0)**.

*This could suggest that Labour voters have more consistent or distinguishable patterns in the dataset.*

### k - Nearest Neibour (kNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
accuracy_scores = []
```

```python
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5)  # 5-fold cross-validation
    accuracy_scores.append(scores.mean())
```

```python
# Plot
plt.plot(range(1, 21), accuracy_scores)
plt.xlabel('n_neighbors')
plt.ylabel('CV Accuracy')
plt.title('KNN Accuracy vs K')
plt.show()
```

```python
# Train a KNN model
best_k = accuracy_scores.index(max(accuracy_scores)) + 1
knn = KNeighborsClassifier(n_neighbors=best_k)  # Default is k=5
knn.fit(X_train, y_train)
```

```python
# Predictions
y_train_pred_knn = knn.predict(X_train)
y_test_pred_knn = knn.predict(X_test)
```

```python
# Evaluation
train_accuracy_knn = accuracy_score(y_train, y_train_pred_knn)
test_accuracy_knn = accuracy_score(y_test, y_test_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_test_pred_knn)
report_knn = classification_report(y_test, y_test_pred_knn)

print("\nTraining Accuracy : ",train_accuracy_knn,'\n')
print("\nAccuracy of Model : ",test_accuracy_knn,'\n')
print("\nConfusion Matrix : \n",conf_matrix_knn,'\n')
print("\nClassification Report: \n ",report_knn,'\n')
```

✅ Overall Model Performance
Training Accuracy: 85.2%

Testing Accuracy: 83.6%

🔎 This small difference between training and test accuracy indicates:

 - The model generalizes well to unseen data.

 - There is no significant overfitting.

Confusion Matrix :

 - True Positives (288): Correctly predicted Labour voters

 - True Negatives (93): Correctly predicted Conservative voters

 - False Positives (45): Predicted Labour but actually Conservative

 - False Negatives (30): Predicted Conservative but actually Labour

Final Inference:

✅ The kNN model is performing strongly overall, especially on classifying Labour voters.

❗ However, it shows some difficulty in identifying Conservative voters, possibly due to:

Less distinctive patterns in their responses.

Class imbalance (fewer Conservative voters).
